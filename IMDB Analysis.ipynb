# Classify Moviee based on the average[Excellent,Good,Average]


```python
df.columns
```

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
df["Rating_cat"]=df['Rating'].apply(rating)
```

# Find Unique Value From Genre

```python
df['Genre']
```

```python
#convert value into list
list1=[]
for value in df['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
#coverting 2D list into 1D list
one_d=[]
for i in list1:
    for j in i:
        one_d.append(j)
```

```python
unique_list=[]
for item in one_d:
    if item not in unique_list:
        unique_list.append(item)
```

```python
unique_list
```
